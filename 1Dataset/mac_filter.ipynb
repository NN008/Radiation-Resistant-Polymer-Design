```python
import pandas as pd

# Load your dataset (with MAC and Tg)
df = pd.read_csv("PI1M_with_Tg_and_MAC.csv")

# Count how many have both Tg and MAC
valid = df.dropna(subset=["Tg", "MAC"])
print(f"✅ Entries with both Tg and MAC: {len(valid)}")

# Count how many were dropped
dropped = len(df) - len(valid)
print(f"❌ Dropped entries with missing Tg or MAC: {dropped}")

# Save the cleaned dataset
valid.to_csv("PI1M_with_Tg_and_MAC_filtered.csv", index=False)
print("💾 Saved to: PI1M_with_Tg_and_MAC_filtered.csv")
```
